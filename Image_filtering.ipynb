# Image Filtering

```python
## importing libraries
import cv2
```

```python
image=cv2.imread("cute_puppy.jpg")
```

#### Gaussian Blur

```python
blurred=cv2.GaussianBlur(image,(11,11),3)
```

```python
cv2.imshow("original",image)    
cv2.imshow("blurred",blurred)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### Median Blur

```python
img=cv2.imread("blackdot.jpg")
```

```python
blurred1=cv2.medianBlur(img,21)
cv2.imshow("original",img)
cv2.imshow("blurred1",blurred1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### Sharpening

```python
## averaging
img2=cv2.imread("low_resolution.jpg")
```

```python
import numpy as np
sharpening_kernel=np.array([
    [0, -1, 0],
    [-1, 5, -1],
    [0, -1, 0]
])

sharpened=cv2.filter2D(img2,-1,sharpening_kernel)
```

```python
cv2.imshow("original", img2)
cv2.imshow("sharpened", sharpened)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
