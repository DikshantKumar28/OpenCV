## Contours

```python
import cv2 
import numpy as np
```

```python
img=cv2.imread("shape.png")
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
thresh=cv2.threshold(gray,127,255,cv2.THRESH_BINARY)
```

```python
contours,hierarchy=cv2.findContours(thresh[1],cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
```

```python
cv2.drawContours(img,contours,-1,(0,255,0),3)
```

```python
cv2.imshow("Contours",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
for countor in contours:
    # area=cv2.contourArea(countor)
    # print(area)
    approx=cv2.approxPolyDP(countor,0.01*cv2.arcLength(countor,True),True)
    corners=len(approx)
    if corners==3:
        shape_name="Triangle"
    elif corners==4:
        shape_name="Rectangle" 
    elif corners>5:
        shape_name="Circle"
    else:
        shape_name="unknown"

    cv2.drawContours(img,[approx],0,(0,255,0),3)
    x=approx.ravel()[0] 
    y=approx.ravel()[1] - 10
    cv2.putText(img,shape_name,(x,y),cv2.FONT_HERSHEY_SIMPLEX,0.5,(255,0,0),1)  
```

```python
cv2.imshow("countors",img)  
cv2.waitKey(0)
cv2.destroyAllWindows()
```
