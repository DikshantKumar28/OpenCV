# Face Detection using HaarCascades

```python
import numpy as np
import cv2
```

```python
face_cascade=cv2.CascadeClassifier("Harcascades/haarcascade_frontalface_default.xml")
eye_cascade=cv2.CascadeClassifier("Harcascades/haarcascade_eye.xml")
smile_cascade=cv2.CascadeClassifier("Harcascades/haarcascade_smile.xml")

for name, cascade in {
    "face": face_cascade,
    "eye": eye_cascade,
    "smile": smile_cascade,
}.items():
    if cascade.empty():
        raise FileNotFoundError(f"Could not load {name} cascade XML file")
```

```python
cap=cv2.VideoCapture(0)
```

```python
while True:
    ret,frame=cap.read()

    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    detect_face=face_cascade.detectMultiScale(gray,1.3 ,5)


 # x- how far from left 
 # y- how far from top
 # w- width of rectangle
# h- height of rectangle
#x+w- how far from left + width of rectangle
# y+h- how far from top + height of rectangle
    for (x,y,w,h) in detect_face:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)
        roi_gray=gray[y:y+h,x:x+w]
        roi_color=frame[y:y+h,x:x+w]

        detect_eye=eye_cascade.detectMultiScale(roi_gray,1.3,22)
        if len(detect_eye) == 0:
            print("No eyes detected")
        else:
             cv2.putText(frame, "Eyes Detected", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)    
        for (ex,ey,ew,eh) in detect_eye:
            cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)

        detect_smile=smile_cascade.detectMultiScale(roi_gray,1.3,22)

        for (sx,sy,sw,sh) in detect_smile:
            cv2.rectangle(roi_color,(sx,sy),(sx+sw,sy+sh),(0,0,255),2)
    cv2.imshow("Face Detection",frame)
    if cv2.waitKey(1) & 0xFF==ord("q"):
        break   
cap.release()
cv2.destroyAllWindows()
```
