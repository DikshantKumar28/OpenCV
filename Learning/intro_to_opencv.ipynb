# Intro to Open Source Computer Vision

```python
import cv2 as cv
```

#### To load the image

```python
image=cv.imread("grayscale.jpg")
```

#### To display the image 

```python
cv.imshow("Hello OpenCV",image)
cv.waitKey(0)
cv.destroyAllWindows()
```

#### To save the file 

```python
image2=cv.imread("image.png")
```

```python
cv.imwrite("cute_puppy.jpg",image2)
```

#### GrayScale Conversion

```python
gray=cv.cvtColor(image2,cv.COLOR_BGR2GRAY)
```

```python
cv.imshow("hello gray",gray)
cv.waitKey(0)
cv.destroyAllWindows() 
```

#### Resizing an image

```python
image.shape
```

```python
resized=cv.resize(image,(500,500))
```

```python
cv.imshow("resized image",resized)
cv.waitKey(0)
cv.destroyAllWindows()  
```

#### Cropping and Slicing

```python
# normally done as numpy array slicing
cropped=image[50:200,200:400]
cv.imshow("original image",image)
cv.imshow("cropped image",cropped)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
h,w=image.shape[:2]
```

```python
center=(w//2,h//2)
M=cv.getRotationMatrix2D(center,90,1.0)
rotated=cv.warpAffine(image,M,(w,h))

cv.imshow("original image",image)
cv.imshow("rotated image",rotated)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
flipped1=cv.flip(image,0)
flipped2=cv.flip(image,1)
flipped3=cv.flip(image,-1)
cv.imshow("original image",image)
cv.imshow("flipped vertically",flipped1)    
cv.imshow("flipped horizontally",flipped2)
cv.imshow("flipped vertically and horizontally",flipped3)
cv.waitKey(0)
cv.destroyAllWindows()
```
