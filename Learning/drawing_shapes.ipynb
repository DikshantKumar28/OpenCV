### Drawing Line & Shapes

```python
# Importing the library
import cv2 as cv
```

```python
image=cv.imread("image.png")
```

```python
image.shape
```

#### 1. Line

```python
pt1=(100,100)
pt2=(200,200)
color=(0,255,0)
thickness=5
cv.line(image,pt1,pt2,color,thickness)
cv.imshow("Line",image)
cv.waitKey(0)
cv.destroyAllWindows()
```

#### 2. Rectangle

```python
pt1=(100,100)
pt2=(200,200)
color=(0,255,0)
thickness=5
cv.rectangle(image,pt1,pt2,color,thickness)
cv.imshow("Rectangle",image)
cv.waitKey(0)
cv.destroyAllWindows()
```

#### 3. Circle

```python
pt1=(200,100)
pt2=(200,200)
color=(0,255,0)
thickness=5
cv.circle(image,pt1,50,color,thickness)
cv.imshow("Circle",image)
cv.waitKey(0)
cv.destroyAllWindows()
```

#### 4. Text

```python
x=str(input("Enter text to display: "))
cv.putText(image,x,(50,50),cv.FONT_HERSHEY_SIMPLEX,1,(255,0,0),2)
cv.imshow("Text",image)
cv.waitKey(0)
cv.destroyAllWindows()
```
